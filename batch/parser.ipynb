```python
import json
import pandas as pd
import numpy as np
import datetime
import re
```

```python
PATH = "/Users/jon/Desktop/Portfolio/flaskr/static/data/flights2.json"
df = pd.read_json(PATH)
df.head()
df.info()
```

```python
def udf(x):
    return str(x)[:4]
df['date'] = df['date'].apply(udf)
df88 = df.where(df['date']=='1988')
df08 = df.where(df['date'] =='2008')
df08.dropna(inplace=True) 
df88.dropna(inplace=True)
df08 = df08.astype({"UniqueCarrier":"string", "count": "int32", "date": "int32"})
df88 = df88.astype({"UniqueCarrier":"string", "count": "int32", "date": "int32"})
```

```python
WRITE_PATH_1 = "/Users/jon/Desktop/Portfolio/flaskr/static/data/flights88.json"
WRITE_PATH_2 = "/Users/jon/Desktop/Portfolio/flaskr/static/data/flights08.json"
```

```python
df08 = df08.groupby(["UniqueCarrier"])["count"].sum()
df88 = df88.groupby(["UniqueCarrier"])["count"].sum()
```

```python
carriers = {"AS": "Alaska Airlines",
            "G4": "Allegiant Air",
            "AA": "American Airlines",
            "DL": "Delta Airlines",
            "9E": "Endeavor Air",
            "MQ": "Envoy Air",
            "F9": "Frontier Airlines",
            "HA": "Hawaiian Airlines",
            "QX": "Horizon Air",
            "B6": "JetBlue Airwys",
            "YV": "Mesa Airlines",
            "OH": "PSA Airlines",
            "YX": "Republic Airlines",
            "OO": "SkyWest Airlines",
            "WN": "Southwest Airlines",
            "NK": "Spirit Airlines",
            "UA": "United Airlines",
            "CO": "Continental Airlines",
            "EA": "Eastern Airlines",
            "HP": "America West Airlines",
            "NW": "Northwest Airlines",
            "PA (1)": "Pan America World Airways",
            "PI": "Piedmont Aviation",
            "PS": "Ukraine International Airlines",
            "TW": "Trans World Airways",
            "US": "US Airways"
            }

dfcarriers = pd.DataFrame.from_dict({"Code":carriers.keys(), "Name":carriers.values()})
```

```python
dfcarriers.head()
joined88 = pd.merge(
    df88,
    dfcarriers,
    how="inner",
    #on=None,
    left_on="UniqueCarrier",
    right_on="Code",
    #left_index=False,
    #right_index=False,
    sort=True,
    #suffixes=("_x", "_y"),
    copy=True,
    indicator=False,
    validate=None,
)
joined08 = pd.merge(
    df08,
    dfcarriers,
    how="inner",
    #on=None,
    left_on="UniqueCarrier",
    right_on="Code",
    #left_index=False,
    #right_index=False,
    sort=True,
    #suffixes=("_x", "_y"),
    copy=True,
    indicator=False,
    validate=None,
)
```
